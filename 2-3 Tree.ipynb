Mr.Code memiliki banyak data dan akan terus bertambah. Sehingga Mr.Code membutuhkan suatu penyimpanan data yang dapat terus menambahkan data dan membuatnya terurut dengan mengimplementasikan metode 2-3 Tree. Penyimpanan data yang dia harapkan dapat memuat operasi penambahan data, pencarian data yang dibutuhkan, dan menampilkan semua data yang dimiliki secara terurut berdasarkan 2-3 Tree. Terkadang Mr.Code hanya ingin membuka data pada 2 level teratas dari tree, maka program juga harus memuat operasi display untuk 2 level teratas dari tree. Sebagai tambahan, Mr.Code yang telah selesai membuka program harus langsung menutupnya supaya program tidak terus berjalan, maka dibutuhkan operasi quit untuk menghentikan program tersebut.

Misal Mr.Code memiliki data 24, 67, 18, 31, 99, 10, 3, 85, 93, 60, 43.
Buatlah operasi untuk menambahkan seluruh data, lalu cari lah data dengan nilai 18 dan 21, print 2 level teratas dari tree, lalu tampilkan seluruh data terurut pada penyimpanan tree tersebut. Terakhir hentikan program.

```python
class Node:
    def __init__(self, data, par = None):
        self.data = list([data])
        self.parent = par
        self.child = list()

    def __str__(self):
        if self.parent:
            return str(self.parent.data) + ' : ' + str(self.data)
        return 'Root : ' + str(self.data)

    def __lt__(self, node):
        return self.data[0] < node.data[0]

    def _isLeaf(self):
        return len(self.child) == 0

    def _add(self, new_node):
        for child in new_node.child:
            child.parent = self
        self.data.extend(new_node.data)
        self.data.sort()
        self.child.extend(new_node.child)
        if len(self.child) > 1:
            self.child.sort()
        if len(self.data) > 2:
            self._split()

    def _insert(self, new_node):

        if self._isLeaf():
            self._add(new_node)

        elif new_node.data[0] > self.data[-1]:
            self.child[-1]._insert(new_node)
        else:
            for i in range(0, len(self.data)):
                if new_node.data[0] < self.data[i]:
                    self.child[i]._insert(new_node)
                    break

    def _split(self):
        left_child = Node(self.data[0], self)
        right_child = Node(self.data[2], self)
        if self.child:
            self.child[0].parent = left_child
            self.child[1].parent = left_child
            self.child[2].parent = right_child
            self.child[3].parent = right_child
            left_child.child = [self.child[0], self.child[1]]
            right_child.child = [self.child[2], self.child[3]]

        self.child = [left_child]
        self.child.append(right_child)
        self.data = [self.data[1]]

        if self.parent:
            if self in self.parent.child:
                self.parent.child.remove(self)
            self.parent._add(self)
        else:
            left_child.parent = self
            right_child.parent = self

    def _find(self, item):
        if item in self.data:
            return item
        elif self._isLeaf():
            return False
        elif item > self.data[-1]:
            return self.child[-1]._find(item)
        else:
            for i in range(len(self.data)):
                if item < self.data[i]:
                    return self.child[i]._find(item)
                
    def _preorder(self):
        print (self) 
        for child in self.child:
            child._preorder()

class Tree:
    def __init__(self):
        self.root = None

    def insert(self, item):
        if self.root is None:
            self.root = Node(item)
        else:
            self.root._insert(Node(item))
            while self.root.parent:
                self.root = self.root.parent
        return True

    def find(self, item):
        return self.root._find(item)

    def printTop2Tiers(self):
        print ('----Top 2 Tiers----')
        print (str(self.root.data))
        for child in self.root.child:
            print (str(child.data), end=' ')
        print(' ')

    def preorder(self):
        print ('-----Preorder-----')
        self.root._preorder()

def main():
    tree = Tree()
    print('Silahkan pilih operasi! (Ingat bahwa key harus selalu berbeda)')
    print('Ketik "add <data>" untuk menambahkan data pada tree')
    print('Ketik "find <data>" untuk melihat apakah data ada atau tidak')
    print('Ketik "preorder" untuk menampilkan tree terurut')
    print('Ketik "display" untuk menampilkan 2 tingkat teratas 2-3 tree')
    print('Ketik "quit" untuk menghentikan program')
    print('=========================================================================')
    while True:
        do = input('Operasi apa yang ingin diproses? ').split()
        operation = do[0].strip().lower()
        if operation == 'add':
            new_data = int(do[1])
            tree.insert(new_data)
        elif operation == 'find':
            new_data = int(do[1])
            search = tree.find(new_data)
            if new_data == search:
                print(search, "ada")
            else:
                print("Data tidak ada")
        elif operation == 'preorder':
            if tree is not None:
                tree.preorder()
            else:
                print('Tree kosong.')
            print('=========================================================================')
        elif operation == 'display':
            if tree is not None:
                tree.printTop2Tiers()
            else:
                print('Tree kosong.')
            print('=========================================================================')
        elif operation == 'quit':
            break
main()
```
